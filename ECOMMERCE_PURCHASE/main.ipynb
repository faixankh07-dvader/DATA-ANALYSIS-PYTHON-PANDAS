5 Highest & Lowest Purchase Price / Average Purchase Price

```python
c = data.columns
highest = data['Purchase Price'].max()
lowest = data['Purchase Price'].min()
average = data['Purchase Price'].mean()
purchase_stats = data["Purchase Price"].describe()

print(c)
print("Highest Purchase Price:", highest)
print("Lowest Purchase Price:", lowest)
print("Average Purchase Price:", average)
print("Purchase Price Stats:\n", purchase_stats)
```

7 Job Analysis

```python
lawyers_users = data[data["Job"].str.contains("lawyer", case=False)]
lawyers_num = len(data[data["Job"].str.contains("lawyer", case=False)])

analyst = len(data[data["Job"].str.contains("Analyst", case=False)])
print(analyst)

print(lawyers_num)

job_count = data["Job"].value_counts()

print(job_count)
lawyers_users
#visualization of jobs ( comparing analyst , engineer and manager)

job_keywords = ["Analyst", "Engineer", "Manager"]

job_counts = {}
for job in job_keywords:
    count = data["Job"].str.contains(job, case=False, na=False).sum()
    job_counts[job] = count

job_df = pd.DataFrame(
    list(job_counts.items()),
         columns=["Job Type", "Count"]
)
job_df.plot(
    x="Job Type",
    y="Count",
    kind="bar",
    legend=False,
    figsize=(8,5)
)
```
